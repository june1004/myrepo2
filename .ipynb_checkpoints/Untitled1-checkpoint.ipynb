## 시각화 라이브러리
 * matplotlib
 * seaborn

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,14,100)    # linspace(시작값, 끝값, 갯수)
x
```

```python
y = np.sin(x)   # 100개의 x 값에 대하여 100개의 sin 값을 구해서 y로 넣어준다....
plt.plot(x, y)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기

plt.plot(x,y1, x,y2, x,y3)
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기

plt.plot(x,y1, label='sin(x)')
plt.plot(x,y2, label='sin(x)*2')
plt.plot(x,y3, label='sin(x)*4')
plt.legend()
plt.title("sin plot")
```

```python
y1 = np.cos(x)
y2 = np.cos(x) * 2
y3 = np.cos(x) * 4

# 그래프 그리기

plt.plot(x,y1, label='cos(x)')
plt.plot(x,y2, label='cos(x)*2')
plt.plot(x,y3, label='cos(x)*4')
plt.legend()
plt.title("cos plot")
```

```python
y1 = np.sin(x)
y2 = np.sin(x) * 2
y3 = np.sin(x) * 4

# 그래프 그리기

plt.plot(x,y1, 'o', label='sin(x)')
plt.plot(x,y2, 'v', label='sin(x)*2')
plt.plot(x,y3, '^', label='sin(x)*4')
plt.legend()
plt.title("sin plot")
```

### 데이터를 불러와서 이에 대해서 간단한 시각화릉 통해 알아보기

```python
import seaborn as sns
```

```python
tips = sns.load_dataset("tips")     # 데이터셋 로드시 판다스 형태로 가져온다...
tips
```

```python
type(tips)     #DataFrame : 2차원적 데이터... CSV 등
```

```python
* [데이터].info()  : 데이터정보
* [데이터].describe()  : 수치형 데이터 요약값
* [데이터].shape   : 행, 열의 정보
* [데이터].columns  : 컬럼명, 제목
* [데이터].head()  : 앞에서 기본적인 데이터 몇개(기본 5개)를 확인해 줌
* [데이터].tail()  : 뒤에서 기본적인 데이터 몇개(기본 5개)를 확인해 줌
* [데이터].isnull().sum()  : 각 열의 결측치(비어있는 값)의 개수.
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("Tips by Day")
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', data=tips)
```

```python
sns.lineplot(x='day', y='total_bill', data=tips)
```

```python
# Initialize a grid of plots tip an Axes for each walk
grid = sns.FacetGrid(tips, col="size", hue="size", palette="tab20c",
                     col_wrap=4, height=1.5)

# Draw a horizontal line to show the starting point
grid.map(plt.axhline, y=0, ls=":", c=".5")

# Draw a line plot to show the trajectory of each random walk
grid.map(plt.plot, "tip", "total_bill", marker="o")

# Adjust the tick positions and labels
grid.set(xticks=np.arange(5), yticks=[-3, 3],
         xlim=(-.5, 4.5), ylim=(-3.5, 3.5))
```
