### pandas로 가져와서.
 * head(), tail() : 앞에서, 또는 뒤에서 일정개수의 데이터 확인
 * info() : 정보를 표시 (행이름, 열이름, 데이터 수, 열의 자료형..)
 * describe() : 몇개의 데이터가 있는지, 평균, 표준편차, 사분위수, 최소, 최대값
 * columns  : 열이름 확인하기 
 * isnull().sum()  : 결측치 개수 확인

```python
tips = sns.load_dataset("tips")
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='total_bill', data=tips)
plt.title("total bills versus days")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='total_bill',hue='smoker'  , data=tips)
plt.title("total bills versus days")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.barplot(x='day', y='tip',hue='sex'  , data=tips)
plt.title("total bills versus days")
plt.show()
```
